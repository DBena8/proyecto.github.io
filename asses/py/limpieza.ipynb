```python
# librerias
import pandas as pd
```

### Equipos

```python
df = pd.read_csv('../../Datasett/equipos.csv')
df.head()
```

```python
# CAMBIAR NOMBRE DE COLUMNA
df.rename(columns={'Franchise': 'Equipos'}, inplace=True)
df.head()
```

```python
df.rename(columns={'Champ': 'Campeonatos'}, inplace=True)
df.head()
```

```python
df.drop(columns=["Lg",'From', "To"], inplace=True)
df.head()
```

#### Graficos

```python
# grafico que muestra los equipos y sus champ 
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df.plot(kind='bar', x='Equipos', y='Campeonatos', color='blue', title='Equipos y sus campeonatos')
```
